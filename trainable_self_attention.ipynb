```python
import torch
```

```python
inputs=torch.tensor([
    [0.43,0.15,0.89], # your
    [0.55,0.87,0.66], # journey
    [0.57,0.85,0.64], # starts
    [0.22,0.58,0.23], # with
    [0.77,0.25,0.10], # one
    [0.05,0.80,0.55]  # step
])
```

```python
d_int=inputs.shape[1]# initailization for shape of query key and value
d_out=2
```

```python
W_query=torch.nn.Parameter(torch.rand(d_int,d_out),requires_grad=False)
W_key=torch.nn.Parameter(torch.rand(d_int,d_out),requires_grad=False)
W_value=torch.nn.Parameter(torch.rand(d_int,d_out),requires_grad=False)
```

```python
query= inputs @ W_query
key=inputs @ W_key
value= inputs @ W_value
```

```python
print(f" Query shape: {query.shape}")
print(f" Key shape: {key.shape}")
print(f" Value shape: {value.shape}")
```

```python
print(value)
```

```python
Attention_score= query @ key.T # Attention_score(6,6)=query(6,2) * key.T(2,6)
```

```python
print(Attention_score)
```

```python
Attention_score=Attention_score/(key.shape[-1]**0.5) # Scaled by root  d , dimension of key 
Attention_score=torch.softmax(Attention_score,dim=-1)
```

```python
class self_attention_v1(torch.nn.Module):
    def __init__(self,d_in,d_out):
        super().__init__()
        self.W_key=torch.nn.Parameter(torch.rand(d_in,d_out))
        self.W_query=torch.nn.Parameter(torch.rand(d_in,d_out))
        self.W_value=torch.nn.Parameter(torch.rand(d_in,d_out))
    
    def forward(self,inputs):
        key=inputs @ self.W_key
        query=inputs @ self.W_query
        value=inputs @ self.W_value
        
        attn_score= query @ key.T
        scaled_attn_score=attn_score/(key.shape[-1]**0.5)
        after_normalized=torch.softmax(scaled_attn_score,dim=-1)
        context_vectors=after_normalized @ value
        return context_vectors
        
```
