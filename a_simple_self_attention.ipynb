```python
import torch
```

```python
inputs=torch.tensor([
    [0.43,0.15,0.89], # your
    [0.55,0.87,0.66], # journey
    [0.57,0.85,0.64], # starts
    [0.22,0.58,0.23], # with
    [0.77,0.25,0.10], # one
    [0.05,0.80,0.55]  # step
])
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
words=["your","journey","starts","with","one","step"]
x_coords=inputs[:,0].numpy()
y_coords=inputs[:,1].numpy()
z_coords=inputs[:,2].numpy()
```

```python
fig=plt.figure()
ax = fig.add_subplot(111,projection='3d')

for x,y,z ,word in zip(z_coords,y_coords,z_coords,words):
    ax.scatter(x,y,z)
    ax.text(x,y,z,word,fontsize=10)

ax.set_xlabel("x")
ax.set_ylabel("y")
ax.set_zlabel("z")
plt.title("3D plot of embeddings")
plt.show()
```

```python
fig=plt.figure()
ax= fig.add_subplot(111,projection='3d')
colors=['r','g','b','c','m','y']

for x,y,z,word,color in zip(x_coords,y_coords,z_coords,words,colors):
    ax.quiver(0,0,0,x,y,z,color=color,arrow_length_ratio=0.05)
    ax.text(x,y,z,word,fontsize=10,color=color)
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.set_zlabel("z")
plt.title("3D plot of embeddings")
ax.set_xlim([0,1])
ax.set_ylim([0,1])
ax.set_zlim([0,1])
plt.show()
```

```python

query=inputs[1]
attention_score_2=torch.empty(inputs.shape[0])
for i, x_i in enumerate(inputs):
    attention_score_2[i]=torch.dot(query,x_i)
    
print(attention_score_2)
```

```python
# now we use normalization for stable training and interpratibility
attention_score_2= attention_score_2/attention_score_2.sum()
print(attention_score_2.sum())
```

```python
# for normailzation softmax is genrally used as e^x-max/sum
normalize=torch.softmax(attention_score_2,dim=0)
print(normalize.sum())
```

```python
query=inputs[1]
context_vec_2=torch.zeros(query.shape)
for i, x_i in enumerate(inputs):
    context_vec_2 += attention_score_2[i]*x_i
```

```python
print(context_vec_2)
```
